```python
#8 Bar graph for People killed in each year and a particular day
```

```python
attack_year=terrors[terrors['iyear']>1900][['iyear','nkill']].groupby('iyear',as_index=False).sum()
attack_month=terrors[terrors['imonth']>0][['imonth','nkill']].groupby('imonth',as_index=False).sum()
attack_day=terrors[terrors['iday']>0][['iday','nkill']].groupby('iday',as_index=False).sum() 
attack_latitude=terrors[terrors['latitude']>0][['latitude','nkill']].groupby('latitude',as_index=False).sum()
```

```python
bar_paint=attack_year.sort_values(by='nkill')
bar_paint2=attack_month.sort_values(by='nkill')
bar_paint3=attack_day.sort_values(by='nkill')
bar_paint4=attack_latitude.sort_values(by='nkill')
```

```python
fig,ax=plt.subplots(figsize=(12,9))
ax.barh(bar_paint['iyear'][::-1],bar_paint['nkill'][::-1])
plt.xticks(rotation=-45)
ax.set_ylabel('Year',size=16)
ax.set_xlabel('People Killed in attacks',size=16)
plt.title('Years',fontsize=20,pad=10,color=sns.cubehelix_palette(8,start=.5,rot=-.75)[-3])
plt.show()




fig,ax=plt.subplots(figsize=(12,6))
#fig,ax=plt.subplots(1,2,figsize=(16,4))
ax.barh(bar_paint2['imonth'][::-1],bar_paint2['nkill'][::-1])
plt.xticks(rotation=-45)
ax.set_ylabel('Month',size=16)
ax.set_xlabel('People Killed in attacks',size=20)
plt.title('Month',fontsize=20,pad=10,color=sns.cubehelix_palette(8,start=.5,rot=-.75)[-3])
plt.show()



```

```python

fig,ax=plt.subplots(figsize=(14,8))
#fig,ax=plt.subplots(1,2,figsize=(16,4))
ax.barh(bar_paint3['iday'][::-1],bar_paint3['nkill'][::-1])
plt.xticks(rotation=-45)
ax.set_ylabel('Day',size=16)
ax.set_xlabel('Number of people Killed ',size=20)
plt.title('People Killed on which date',fontsize=20)
plt.show()
```
